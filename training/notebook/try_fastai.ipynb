```python
import os
stem_by_lbl = {l.stem: l for l in (p / "labels").glob("*.txt")}
pos= {
    k: v
    for (k,v) in stem_by_lbl.items()
    if os.path.getsize(v) > 0
}
negs = {
    k: v
    for (k,v) in stem_by_lbl.items()
    if k not in pos
}
# make it smaller
negs = {
    k: negs[k]
    for k in list(negs.keys())[:len(pos)*2]
}
len(pos), len(negs)
```

```python
from fastai.data.all import *
from fastai.vision.all import *
```

```python

def _data_loader_iter(self):
    for batch in map(self.get_batch, iter(self.batch_sampler)):
        yield get_tensor(batch, self.pin_memory)
DataLoader.__iter__ = _data_loader_iter
```

```python
def _get_items(base_path):
    def _in_lbls(img_path):
        lbl_name = img_path.stem
        return lbl_name in pos or lbl_name in negs
    return list(filter(_in_lbls, get_image_files(base_path)))


def _get_y(img_path):
    b = img_path.stem
    if b in pos:
        return "y"
    else:
        return "n"
```

```python
dblock = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=_get_items,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=_get_y,
    item_tfms=Resize(128),
)
```

```python
dload = dblock.dataloaders(images_base, batch_size=1, num_workers=1)
```
