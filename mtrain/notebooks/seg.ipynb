```python
from ultralytics import YOLO
model = YOLO("yolov8n-seg.pt")
```

```python
# Run segmentation
# img_path = ""
p = "/Users/hariomnarang/Desktop/personal/roads/datasets/samples_mapillary/ahmedabad/images/252987033288051.jpg" 
results = model.predict(p)

# results[0].masks contains segmentation masks
results[0].show()  # visualizes the mask overlay
```

```python
# COCO classes that might represent ground
ground_classes = ["road", "sidewalk", "grass", "pavement"]  

for result in results:
    masks = result.masks
    cls_ids = result.boxes.cls  # class indices predicted
    
    for i, cls_id in enumerate(cls_ids):
        cls_name = model.names[int(cls_id)]
        print(cls_name)
        if cls_name in ground_classes:
            masks.data[i].show()  # shows only ground mask
```
