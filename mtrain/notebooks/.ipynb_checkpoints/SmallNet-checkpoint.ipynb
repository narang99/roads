```python
import numpy as np
from typing import List, Tuple

def split_image_into_tiles(
    img: np.ndarray,
    tile_size: int = 50
) -> List[Tuple[np.ndarray, Tuple[int, int]]]:
    H, W = img.shape[:2]
    tiles = []

    for y in range(0, H, tile_size):
        for x in range(0, W, tile_size):
            tile = img[y:y + tile_size, x:x + tile_size]
            tiles.append((tile, (y, x)))

    return tiles

import cv2
import numpy as np
from typing import List, Tuple

def highlight_tiles_bbox(
    img: np.ndarray,
    tiles: List[Tuple[np.ndarray, Tuple[int, int]]],
    positives: List[bool],
    tile_size: int = 50,
    color=(0, 255, 0),
    thickness=2
) -> np.ndarray:
    img_out = img.copy()
    H, W = img.shape[:2]

    for (_, (y, x)), is_pos in zip(tiles, positives):
        if not is_pos:
            continue

        y2 = min(y + tile_size, H)
        x2 = min(x + tile_size, W)

        cv2.rectangle(
            img_out,
            (x, y),
            (x2, y2),
            color,
            thickness
        )

    return img_out
```
